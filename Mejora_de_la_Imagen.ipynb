#Mejora de la Imagen


El procesamiento de imágenes biomédicas, incluyendo la mejora de imágenes, es una parte fundamental en la investigación y la práctica médica. Los principales objetivos de la mejora de imágenes biomédicas son:

1. Mejora de la visibilidad de estructuras: A menudo, las imágenes biomédicas pueden contener detalles que son difíciles de distinguir debido al ruido, la baja calidad de imagen o la presencia de artefactos. El objetivo principal de la mejora de imágenes es aumentar la visibilidad de las estructuras de interés para facilitar su análisis y diagnóstico.

2. Reducción del ruido: El ruido puede introducirse en las imágenes biomédicas debido a varios factores, como la adquisición de imágenes en condiciones de poca luz o la interferencia electromagnética. La mejora de imágenes busca reducir o eliminar el ruido para obtener imágenes más claras y precisas.

3. Realce de contraste: Mejorar el contraste de una imagen permite resaltar diferencias en la intensidad de los píxeles, lo que puede revelar detalles importantes que de otra manera serían difíciles de discernir.

4. Corrección de artefactos: Las imágenes biomédicas pueden contener artefactos que distorsionan la información, como el efecto de desplazamiento debido al movimiento del paciente durante la adquisición. La mejora de imágenes puede ayudar a corregir estos artefactos.

5. Normalización y calibración: En algunos casos, es necesario normalizar o calibrar las imágenes biomédicas para que los valores de píxeles se correspondan con unidades físicas conocidas, lo que facilita la medición y comparación de datos.

## Ecualización del histograma.

Un histograma es una representación gráfica de la distribución de las intensidades de píxeles en una imagen. En un histograma, el eje horizontal representa los niveles de intensidad (generalmente en una escala de 0 a 255 para imágenes en escala de grises) y el eje vertical muestra la cantidad de píxeles en la imagen que tienen cada nivel de intensidad. Básicamente, un histograma muestra cuántos píxeles tienen una intensidad particular en la imagen. Esto es útil para comprender la distribución de intensidades y la información de contraste en una imagen.

La ecualización del histograma, por otro lado, es un proceso de mejora de imágenes que se utiliza para mejorar el contraste de una imagen. El objetivo principal de la ecualización del histograma es distribuir las intensidades de píxeles de manera más uniforme en todo el rango de intensidades disponible (por lo general, de 0 a 255) para que haya un uso óptimo de este rango y, como resultado, se resalten mejor los detalles en la imagen. Los pasos para la ecualización del histograma son los siguientes:

1. Calcular el histograma de la imagen original.
2. Calcular la función de transformación acumulativa (CDF, por sus siglas en inglés) del histograma original.
3. Aplicar la CDF a la imagen original para obtener la imagen mejorada.

La ecualización del histograma redistribuirá las intensidades de píxeles de manera que los niveles de intensidad se utilicen de manera más uniforme, lo que a menudo conduce a una mejora en el contraste y la visibilidad de detalles en la imagen.

Este proceso es especialmente útil cuando una imagen tiene un rango de intensidades estrecho o cuando el contraste es bajo. Sin embargo, la ecualización del histograma también puede aumentar el ruido en áreas de baja intensidad y no siempre es adecuada para todas las imágenes.

Es importante mencionar que la ecualización del histograma es una transformación global y puede no ser adecuada para todas las imágenes, ya que puede cambiar drásticamente la apariencia de una imagen. En algunos casos, se pueden utilizar técnicas de mejora de contraste locales para evitar cambios no deseados en áreas específicas de la imagen.

Asegúrate de tener la biblioteca [scikit-image](https://scikit-image.org/docs/stable/auto_examples/data/plot_scientific.html#sphx-glr-auto-examples-data-plot-scientific-py) instalada antes de ejecutar este código:

En este ejemplo, utilizamos la función `equalize_hist` de scikit-image para mejorar el contraste de una imagen de células sanguíneas. La imagen original y la imagen mejorada se muestran lado a lado para ilustrar cómo la mejora de contraste puede resaltar detalles en la imagen.

```python
import matplotlib.pyplot as plt
from skimage import data, exposure

# Cargamos una imagen biomédica de células sanguíneas (debe cambiar la URL por una imagen real)
imagen = data.lily()

# Aplicamos una mejora de contraste utilizando la función 'equalize_hist' de scikit-image
imagen_mejorada = exposure.equalize_hist(imagen)

# Mostramos la imagen original y la imagen mejorada
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Imagen Original")
plt.imshow(imagen, cmap='gray')

plt.subplot(1, 2, 2)
plt.title("Imagen Mejorada")
plt.imshow(imagen_mejorada, cmap='gray')

plt.show()
```

## Filtrado de imagenes para eliminacion de ruido

La reducción de ruido en imágenes es un proceso esencial en el procesamiento de imágenes para mejorar la calidad y la interpretabilidad de las imágenes digitales. Los métodos de filtrado de imágenes se utilizan para eliminar o disminuir el ruido presente en las imágenes, lo que puede mejorar la precisión de las técnicas de análisis y reconocimiento de patrones. Estas técnicas se dividen en dos categorías principales: filtrado de imágenes lineales y no lineales.

**Enumeración de técnicas de filtrado de imágenes para reducir el ruido:**

1. **Filtrado lineal**
2. **Filtrado media (Promedio)**
3. **Filtrado de mediana**
4. **Filtrado de Gauss**
5. **Filtrado bilateral**
6. **Filtrado no lineal de orden estadístico**
7. **Filtrado adaptativo**
8. **Filtrado de restauración**

Ahora, profundicemos en cada una de estas técnicas:

**1. Filtrado lineal:**
   - El filtrado lineal se basa en la convolución de la imagen original con un núcleo (kernel) de filtro lineal predefinido.
   - Ejemplos comunes incluyen el filtro de paso bajo, el filtro de paso alto y el filtro de realce de bordes.

**2. Filtrado media (Promedio):**
   - El filtrado de media se utiliza para suavizar una imagen al reemplazar cada píxel por el valor promedio de los píxeles en su vecindario.
   - Es eficaz para eliminar el ruido de tipo sal y pimienta, pero puede perder detalles finos.

**3. Filtrado de mediana:**
   - El filtrado de mediana es una técnica no lineal que reemplaza cada píxel por el valor mediano de los píxeles en su vecindario.
   - Es efectivo para eliminar ruido impulsivo (sal y pimienta) mientras conserva mejor los detalles que el filtrado de media.

**4. Filtrado de Gauss:**
   - El filtrado de Gauss aplica un filtro basado en la función de Gauss a la imagen, suavizando gradualmente los píxeles.
   - Es eficaz para eliminar el ruido gaussiano y suavizar la imagen.

**5. Filtrado bilateral:**
   - El filtrado bilateral es una técnica no lineal que suaviza la imagen al tiempo que preserva los bordes.
   - Combina el filtrado gaussiano con un componente basado en la diferencia de intensidades entre los píxeles.

**6. Filtrado no lineal de orden estadístico:**
   - Estos filtros se basan en propiedades estadísticas de los píxeles en su vecindario y pueden incluir filtros de mínimo, máximo, rango, entre otros.

**7. Filtrado adaptativo:**
   - El filtrado adaptativo ajusta el filtro según la información local de la imagen.
   - El filtro se adapta a las variaciones locales de contraste para preservar mejor los detalles.

**8. Filtrado de restauración:**
   - El filtrado de restauración se utiliza para recuperar una imagen original a partir de una imagen degradada por ruido y otros factores.
   - Se basa en la aplicación de algoritmos para restaurar la información perdida durante la adquisición o el proceso de degradación de la imagen.

Cada técnica de filtrado tiene sus propias ventajas y desventajas, y la elección de la técnica adecuada dependerá del tipo de ruido presente en la imagen, los detalles que se deben conservar y los requisitos específicos de la aplicación. En la práctica, es común aplicar varias técnicas de filtrado de imágenes en combinación para lograr los mejores resultados de reducción de ruido y mejora de calidad de imagen.

### Filtrado lineal

El filtrado lineal es una técnica común en el procesamiento de imágenes que se basa en la convolución de una imagen con un núcleo de filtro lineal. Matemáticamente, la convolución de una imagen 2D \(I\) con un filtro \(F\) se denota como \(I * F\) y se calcula de la siguiente manera:

$$
(I * F)(x, y) = \sum_{i=-\infty}^{\infty} \sum_{j=-\infty}^{\infty} I(x - i, y - j) \cdot F(i, j)
$$

Donde:
- \(I(x, y)\) es el valor de intensidad del píxel en la posición \((x, y)\) de la imagen original.
- \(F(i, j)\) es el valor del núcleo del filtro en la posición \((i, j)\).
- La suma se realiza sobre todas las posiciones \((i, j)\) del núcleo de filtro que se superponen con la imagen en \((x, y)\).

A continuación, te mostraré ejemplos de filtrado lineal utilizando la imagen "lily" de scikit-image y tres tipos de filtros comunes: pasa bajo, pasa alto y de realce de bordes.

Primero, asegúrate de tener scikit-image instalado y luego sigue este código de ejemplo en Python:

En este ejemplo, primero cargamos la imagen "lily" en escala de grises. Luego, definimos tres filtros diferentes: un filtro pasa bajo (filtro de promedio), un filtro pasa alto (filtro Laplaciano) y un filtro de realce de bordes. Después, aplicamos cada uno de estos filtros a la imagen original utilizando la función `ndimage.convolve` de SciPy. Finalmente, mostramos las imágenes originales y filtradas para visualizar los efectos de cada filtro en la imagen.

```python
import matplotlib.pyplot as plt
from skimage import data, io, color
from scipy import ndimage
import numpy as np

# Cargar la imagen "microanareurysms" de scikit-image
imagen = data.microaneurysms()

# Definir un filtro pasa bajo (filtro de promedio)
filtro_pasa_bajo = np.ones((3, 3)) / 9.0  # Un filtro de promedio 5x5

# Aplicar el filtro pasa bajo a la imagen
imagen_pasa_bajo = ndimage.convolve(imagen, filtro_pasa_bajo, mode='reflect')

# Definir un filtro pasa alto (filtro Laplaciano)
filtro_pasa_alto = np.array([[-1, -1, -1],
                             [-1, 8, -1],
                             [-1, -1, -1]])

# Aplicar el filtro pasa alto a la imagen
imagen_pasa_alto = ndimage.convolve(imagen, filtro_pasa_alto, mode='reflect')

# Definir un filtro de realce de bordes (filtro de realce)
filtro_realce_bordes = np.array([[-1, -1, -1],
                                 [-1, 9, -1],
                                 [-1, -1, -1]])

# Aplicar el filtro de realce de bordes a la imagen
imagen_realce_bordes = ndimage.convolve(imagen, filtro_realce_bordes, mode='reflect')

# Mostrar las imágenes originales y filtradas
plt.figure(figsize=(12, 4))
plt.subplot(141)
plt.title("Imagen Original")
plt.imshow(imagen, cmap='gray')

plt.subplot(142)
plt.title("Filtro Pasa Bajo")
plt.imshow(imagen_pasa_bajo, cmap='gray')

plt.subplot(143)
plt.title("Filtro Pasa Alto")
plt.imshow(imagen_pasa_alto, cmap='gray')

plt.subplot(144)
plt.title("Filtro Realce de Bordes")
plt.imshow(imagen_realce_bordes, cmap='gray')

plt.tight_layout()
plt.show()
```

### Filtrado media

El filtrado de media, también conocido como filtrado promedio, es una técnica de procesamiento de imágenes que se utiliza para suavizar la imagen y reducir el ruido. Esta técnica se basa en el cálculo del valor promedio de los píxeles en un vecindario específico y reemplazar el valor del píxel central por este valor promedio. Matemáticamente, el filtrado de media se expresa de la siguiente manera:

$$
I'(x, y) = \frac{1}{N} \sum_{i=-k}^{k} \sum_{j=-k}^{k} I(x+i, y+j)
$$

Donde:
- \(I'(x, y)\) es el valor del píxel suavizado en la posición \((x, y)\).
- \(I(x+i, y+j)\) son los valores de los píxeles en el vecindario alrededor de \((x, y)\).
- \(N\) es el número total de píxeles en el vecindario (típicamente \(N = (2k + 1)^2\) en un vecindario de tamaño \((2k + 1) \times (2k + 1)\)).

A continuación, se muestra un ejemplo de cómo aplicar el filtrado de media a la imagen "microaneurysms" de scikit-image en Python:


En este código, primero cargamos la imagen "microaneurysms" en escala de grises. Luego, definimos el tamaño del vecindario (kernel) que se utilizará para calcular el valor promedio de los píxeles en ese vecindario. Finalmente, aplicamos el filtrado de media a la imagen utilizando `ndimage.uniform_filter` y mostramos la imagen original y la imagen suavizada resultante.

El filtrado de media es efectivo para suavizar la imagen y reducir el ruido, pero también puede difuminar los detalles finos. El tamaño del vecindario (kernel) determina el grado de suavizado, donde un kernel más grande suavizará más la imagen.

```python
import matplotlib.pyplot as plt
from skimage import data, io, color, img_as_ubyte
from scipy import ndimage
import numpy as np

# Cargar la imagen "microaneurysms" de scikit-image en escala de grises
imagen = data.microaneurysms()

# Definir el tamaño del vecindario (kernel)
k = 3  # Tamaño del vecindario 7x7 (3 píxeles a cada lado)

# Aplicar el filtrado de media a la imagen
imagen_suavizada = ndimage.uniform_filter(imagen, size=(2*k+1, 2*k+1))

# Mostrar las imágenes original y suavizada
plt.figure(figsize=(10, 4))
plt.subplot(121)
plt.title("Imagen Original")
plt.imshow(imagen, cmap='gray')

plt.subplot(122)
plt.title("Imagen Suavizada (Filtrado de Media)")
plt.imshow(imagen_suavizada, cmap='gray')

plt.tight_layout()
plt.show()
```

### Filtro Mediana

El filtrado mediana es otra técnica común utilizada en el procesamiento de imágenes para eliminar el ruido y conservar los detalles de la imagen. A diferencia del filtrado de media, el filtrado mediana calcula el valor del píxel central en función de la mediana de los valores de los píxeles en un vecindario específico. Esto ayuda a eliminar valores extremos o ruidosos sin suavizar demasiado los detalles finos.

Matemáticamente, el filtrado mediana se expresa de la siguiente manera:

$$
I'(x, y) = \text{mediana}\left(\{I(x+i, y+j) : \forall i, j \text{ en el vecindario}\}\right)
$$

A continuación, se muestra un ejemplo de cómo aplicar el filtrado mediana a la imagen "microaneurysms" de scikit-image en Python:

En este código, cargamos la imagen "microaneurysms" en escala de grises y luego aplicamos el filtrado mediana utilizando `ndimage.median_filter`. El tamaño del vecindario (kernel) se define de manera similar al ejemplo anterior. La imagen suavizada resultante conservará mejor los detalles mientras elimina el ruido. El filtrado mediana es especialmente útil cuando se desea preservar características finas en la imagen.

```python
import matplotlib.pyplot as plt
from skimage import data, io, color, img_as_ubyte
from scipy import ndimage
import numpy as np

# Cargar la imagen "microaneurysms" de scikit-image en escala de grises
imagen = data.microaneurysms()

# Definir el tamaño del vecindario (kernel)
k = 3  # Tamaño del vecindario 7x7 (3 píxeles a cada lado)

# Aplicar el filtrado mediana a la imagen
imagen_suavizada = ndimage.median_filter(imagen, size=(2*k+1, 2*k+1))

# Mostrar las imágenes original y suavizada
plt.figure(figsize=(10, 4))
plt.subplot(121)
plt.title("Imagen Original")
plt.imshow(imagen, cmap='gray')

plt.subplot(122)
plt.title("Imagen Suavizada (Filtrado de Mediana)")
plt.imshow(imagen_suavizada, cmap='gray')

plt.tight_layout()
plt.show()
```

### Filtro Gaussiano

El filtrado Gaussiano es una técnica de procesamiento de imágenes que se utiliza para suavizar la imagen y reducir el ruido, similar al filtrado de media. Sin embargo, en lugar de calcular el promedio de los píxeles en un vecindario, el filtrado Gaussiano asigna un mayor peso a los píxeles cercanos al píxel central y un peso menor a los píxeles más alejados. Esto crea un efecto de suavizado que preserva mejor los detalles finos de la imagen.

La operación de convolución en el filtrado Gaussiano se realiza utilizando un núcleo de filtro Gaussiano, que es una función de campana simétrica bidimensional. Matemáticamente, el filtrado Gaussiano se expresa de la siguiente manera:

$$
I'(x, y) = \frac{1}{2\pi\sigma^2} \sum_{i=-k}^{k} \sum_{j=-k}^{k} e^{-\frac{{i^2 + j^2}}{{2\sigma^2}}} \cdot I(x+i, y+j)
$$

Donde:
- $I'(x, y)$ es el valor del píxel suavizado en la posición $(x, y)$.
- $I(x+i, y+j)$ son los valores de los píxeles en el vecindario alrededor de $(x, y)$.
- $\sigma$ es la desviación estándar del núcleo Gaussiano, que controla la cantidad de suavizado.
- $k$ es el tamaño del vecindario, que suele ser un valor positivo entero para definir la ventana del filtro.

A continuación, se muestra un ejemplo de cómo aplicar el filtrado Gaussiano a la imagen "microaneurysms" de scikit-image en Python:

En este código, hemos cargado la imagen "microaneurysms" en escala de grises y aplicado el filtrado Gaussiano utilizando `ndimage.gaussian_filter`. Puedes ajustar la desviación estándar ($\sigma$) para controlar la cantidad de suavizado que se aplica a la imagen. Un valor mayor de $\sigma$ proporcionará un suavizado más suave, mientras que un valor menor de $\sigma$ producirá un suavizado más fuerte. El filtrado Gaussiano es útil cuando se desea un suavizado suave y controlado.

```python
import matplotlib.pyplot as plt
from skimage import data, io, color, img_as_ubyte
from scipy import ndimage
import numpy as np

# Cargar la imagen "microaneurysms" de scikit-image en escala de grises
imagen = data.microaneurysms()

# Definir la desviación estándar del núcleo Gaussiano
sigma = 1.5

# Aplicar el filtrado Gaussiano a la imagen
imagen_suavizada = ndimage.gaussian_filter(imagen, sigma=sigma)

# Mostrar las imágenes original y suavizada
plt.figure(figsize=(10, 4))
plt.subplot(121)
plt.title("Imagen Original")
plt.imshow(imagen, cmap='gray')

plt.subplot(122)
plt.title("Imagen Suavizada (Filtrado Gaussiano)")
plt.imshow(imagen_suavizada, cmap='gray')

plt.tight_layout()
plt.show()
```

### Filtto Bilateral

El filtrado bilateral es una técnica de procesamiento de imágenes que se utiliza para suavizar la imagen y reducir el ruido, al tiempo que preserva los bordes y los detalles importantes. A diferencia del filtrado Gaussiano estándar, el filtrado bilateral tiene en cuenta tanto la proximidad espacial como las diferencias en intensidad de los píxeles al realizar el suavizado. Esto permite que el filtrado bilateral preserve los bordes y los detalles de la imagen al tiempo que reduce el ruido.

Matemáticamente, el filtrado bilateral se expresa de la siguiente manera:

$$
I'(x, y) = \frac{1}{W(x, y)} \sum_{i=-k}^{k} \sum_{j=-k}^{k} w(i, j, \sigma_s) \cdot w(I(x, y) - I(x+i, y+j), \sigma_r) \cdot I(x+i, y+j)
$$

Donde:
- $I'(x, y)$ es el valor del píxel suavizado en la posición $(x, y)$.
- $I(x, y)$ es el valor del píxel en la posición $(x, y)$ en la imagen original.
- $k$ es el tamaño del vecindario, que suele ser un valor positivo entero para definir la ventana del filtro.
- $W(x, y)$ es el peso normalizador que asegura que la suma de los pesos sea igual a 1.
- $w(i, j, \sigma_s)$ es una función de peso espacial que depende de la distancia espacial entre $(x, y)$ y $(x+i, y+j)$ y se controla mediante el parámetro $\sigma_s$, que controla la proximidad espacial.
- $w(I(x, y) - I(x+i, y+j), \sigma_r)$ es una función de peso de rango que depende de la diferencia de intensidad entre los píxeles $(x, y)$ y $(x+i, y+j)$, controlada por el parámetro $\sigma_r$, que controla las diferencias en la intensidad.

A continuación, se muestra un ejemplo de cómo aplicar el filtrado bilateral a la imagen "microaneurysms" de scikit-image en Python:

En este código, hemos cargado la imagen "microaneurysms" en escala de grises y aplicado el filtrado bilateral utilizando la función `denoise_bilateral` de scikit-image. Puedes ajustar los parámetros `sigma_spatial` y `sigma_range` para controlar la proximidad espacial y las diferencias en la intensidad, respectivamente. El filtrado bilateral es útil cuando se desea suavizar la imagen mientras se conservan los detalles y los bordes importantes.

```python
import matplotlib.pyplot as plt
from skimage import data, io, color, img_as_ubyte
from skimage.restoration import denoise_bilateral
import numpy as np

# Cargar la imagen "microaneurysms" de scikit-image en escala de grises
imagen = data.microaneurysms()

# Definir los parámetros del filtrado bilateral
sigma_spatial = 3  # Parámetro sigma para el peso espacial
sigma_range = 0.1  # Parámetro sigma para el peso de rango

# Aplicar el filtrado bilateral a la imagen
imagen_suavizada = denoise_bilateral(imagen, sigma_color=sigma_range, sigma_spatial=sigma_spatial)

# Mostrar las imágenes original y suavizada
plt.figure(figsize=(10, 4))
plt.subplot(121)
plt.title("Imagen Original")
plt.imshow(imagen, cmap='gray')

plt.subplot(122)
plt.title("Imagen Suavizada (Filtrado Bilateral)")
plt.imshow(imagen_suavizada, cmap='gray')

plt.tight_layout()
plt.show()
```

### Filtrado de orden estadístico (Moda)

Si estás buscando un filtro no lineal de orden estadístico que no sea la mediana, uno de los ejemplos más comunes es el filtro de moda. El filtro de moda se utiliza para encontrar el valor más común en un vecindario y se utiliza para eliminar ruido en imágenes mientras se preservan los detalles.

Matemáticamente, el filtro de moda se expresa de la siguiente manera:

$$
I'(x, y) = \text{moda}\left(\{I(x+i, y+j) : \forall i, j \text{ en el vecindario}\}\right)
$$

Donde:
- $I'(x, y)$ es el valor del píxel suavizado en la posición $(x, y)$.
- $I(x+i, y+j)$ son los valores de los píxeles en el vecindario alrededor de $(x, y)$.

A continuación, se muestra un ejemplo de cómo aplicar el filtro de moda a la imagen "microaneurysms" de scikit-image en Python. En este código, hemos cargado la imagen "microaneurysms" en escala de grises y aplicado el filtro de moda utilizando la función `ndimage.generic_filter` junto con la función `local_mode` que calcula la moda en un vecindario. Puedes ajustar el tamaño del vecindario (`k`) según tus necesidades. El filtro de moda es eficaz para eliminar el ruido sin suavizar en exceso los detalles importantes de la imagen.


Claro, la función `local_mode` es una función personalizada que se utiliza para calcular la moda (valor más común) de un conjunto de valores en un vecindario específico. Aquí está una descripción de lo que hace cada parte de la función:

1. `values, counts = np.unique(arr, return_counts=True)`: Esta línea utiliza la función `np.unique` de NumPy para encontrar los valores únicos en el array `arr` y contar cuántas veces aparece cada valor. `values` es un array que contiene los valores únicos en `arr`, y `counts` es un array que contiene la cantidad de veces que aparece cada valor en el mismo orden que en `values`.

2. `mode_value = values[np.argmax(counts)]`: Después de encontrar los valores únicos y sus recuentos, esta línea utiliza `np.argmax` para encontrar la posición del valor que tiene el recuento máximo en el array `counts`. Luego, se utiliza esta posición para obtener el valor correspondiente en el array `values`, que es la moda de `arr`.

3. `return mode_value`: Finalmente, la función devuelve la moda calculada como `mode_value`.

En resumen, esta función toma un array de valores (como un vecindario de píxeles en una imagen), encuentra los valores únicos en ese vecindario, cuenta cuántas veces aparece cada valor y devuelve el valor que aparece con mayor frecuencia, es decir, la moda. Esto es útil para aplicar el filtro de moda en procesamiento de imágenes, donde se desea reemplazar un píxel por el valor más común en su vecindario para eliminar el ruido.

```python
import matplotlib.pyplot as plt
from skimage import data, io, color, img_as_ubyte
from scipy import ndimage
import numpy as np
from scipy import stats

# Cargar la imagen "microaneurysms" de scikit-image en escala de grises
imagen = data.microaneurysms()

# Definir el tamaño del vecindario (kernel)
k = 3  # Tamaño del vecindario 7x7 (3 píxeles a cada lado)

# Crear una función para calcular la moda de un vecindario
def local_mode(arr):
    values, counts = np.unique(arr, return_counts=True)
    mode_value = values[np.argmax(counts)]
    return mode_value

# Aplicar el filtro de moda a la imagen
imagen_suavizada = ndimage.generic_filter(imagen, local_mode, size=(2*k+1, 2*k+1))

# Mostrar las imágenes original y suavizada
plt.figure(figsize=(10, 4))
plt.subplot(121)
plt.title("Imagen Original")
plt.imshow(imagen, cmap='gray')

plt.subplot(122)
plt.title("Imagen Suavizada (Filtro de Moda)")
plt.imshow(imagen_suavizada, cmap='gray')

plt.tight_layout()
plt.show()
```

### Filtros Adaptativos

Los filtros adaptativos son técnicas de procesamiento de imágenes que ajustan su comportamiento según las características locales de la imagen. Los dos tipos de filtros adaptativos más comunes son:

1. **Filtro de Mediana Adaptativo**: Este filtro se basa en el filtro de mediana estándar, pero en lugar de aplicar una ventana de mediana fija, ajusta el tamaño de la ventana de mediana en función de la variabilidad local de la imagen. Es útil para eliminar el ruido en imágenes con diferentes niveles de detalle.

2. **Filtro de Media Geométrica Adaptativo**: Este filtro utiliza una ventana de tamaño variable y calcula la media geométrica de los píxeles dentro de la ventana. La media geométrica es menos sensible a los valores extremos en comparación con la media aritmética y puede preservar mejor los detalles de la imagen.

A continuación, se muestra el código para aplicar estos dos tipos de filtros adaptativos a una imagen utilizando Python y scikit-image. Asegúrate de ajustar los parámetros según tus necesidades:


```python
import matplotlib.pyplot as plt
from skimage import data, io, color, img_as_ubyte
from scipy import ndimage
import numpy as np

# Cargar la imagen "microaneurysms" de scikit-image en escala de grises
imagen = data.microaneurysms()

# Definir el tamaño de la ventana adaptativa
window_size = 3  # Ajusta el tamaño según sea necesario

# Aplicar el filtro de mediana adaptativo a la imagen
imagen_suavizada_mediana_adaptativa = ndimage.median_filter(imagen, size=window_size)

# Mostrar las imágenes original y suavizada
plt.figure(figsize=(10, 4))
plt.subplot(121)
plt.title("Imagen Original")
plt.imshow(imagen, cmap='gray')

plt.subplot(122)
plt.title("Imagen Suavizada (Filtro de Mediana Adaptativo)")
plt.imshow(imagen_suavizada_mediana_adaptativa, cmap='gray')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from skimage import data, io, color, img_as_ubyte
from scipy import ndimage
import numpy as np

# Cargar la imagen "microaneurysms" de scikit-image en escala de grises
imagen = data.microaneurysms()

# Definir el tamaño de la ventana adaptativa
window_size = 3  # Ajusta el tamaño según sea necesario

# Aplicar el filtro de media geométrica adaptativo a la imagen
def adaptive_geometric_mean(arr):
    return np.exp(np.mean(np.log(arr)))

imagen_suavizada_media_geom_adaptativa = ndimage.generic_filter(imagen, adaptive_geometric_mean, size=(window_size, window_size))

# Mostrar las imágenes original y suavizada
plt.figure(figsize=(10, 4))
plt.subplot(121)
plt.title("Imagen Original")
plt.imshow(imagen, cmap='gray')

plt.subplot(122)
plt.title("Imagen Suavizada (Filtro de Media Geométrica Adaptativo)")
plt.imshow(imagen_suavizada_media_geom_adaptativa, cmap='gray')

plt.tight_layout()
plt.show()
```
